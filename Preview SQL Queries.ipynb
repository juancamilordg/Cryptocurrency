# Preview Databases Ethereum Dataset

```python
# Execute Query to get sum and average transacction gas fees ETH per day
query_string = """
SELECT
  SUM(value/POWER(10,18)) AS sum_tx_ether,
  AVG(gas_price*(receipt_gas_used/POWER(10,18))) AS avg_tx_gas_cost,
  DATE(timestamp) AS tx_date
FROM
  `bigquery-public-data.crypto_ethereum.transactions` AS transactions,
  `bigquery-public-data.crypto_ethereum.blocks` AS blocks
WHERE TRUE
  AND transactions.block_number = blocks.number
  AND receipt_status = 1
  AND value > 0
GROUP BY tx_date
HAVING tx_date >= '2018-01-01' AND tx_date <= '2022-05-31'
ORDER BY tx_date
"""

dataframe = (
    bqclient.query(query_string)
    .result()
    .to_dataframe(
        create_bqstorage_client=True,
    )
)
print(dataframe.head())
```

```python
# Create plot with Average Ether transactions cost by day
plt.style.use('ggplot')
sns.set_context("notebook", font_scale=1.3, rc={"lines.linewidth": 0.5})

f, g = plt.subplots(figsize=(12, 9))
g = sns.lineplot(x="tx_date", y="avg_tx_gas_cost", data=dataframe, palette="Blues_d")
plt.title("Average Ether transaction cost over time")
plt.show(g)
```
